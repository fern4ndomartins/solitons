```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 5, 1000)
colors = ['red', 'green', 'blue', 'purple', 'yellow']
plt.figure(figsize=(9,4.5))
for m in range(5):
    plt.plot(x, np.sin((np.pi*x*(m+1))/5), label=f'mode n={m+1}', color=colors[m])

plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Normal Modes')
plt.legend(loc='lower left', ncol=2)
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

# Parameters
N = 6              # number of masses
radius = 0.2       # radius of each ball
spring_points = 200 # how detailed the spring zig-zag is
a = 2.0            # spacing between masses

# Positions of the masses (x-coordinates)
x = np.arange(N) * a
y = np.zeros(N)

fig, ax = plt.subplots(figsize=(10, 3))

# Draw springs between masses
for i in range(N-1):
    x_start, x_end = x[i] + radius, x[i+1] - radius
    xs = np.linspace(x_start, x_end, spring_points)
    ys = 0.2 * np.sin(5 * np.pi * (xs - x_start) / (x_end - x_start))
    ax.plot(xs, ys, color='black')

# Draw masses (balls) with annotations
for i, (xi, yi) in enumerate(zip(x, y), 1):
    circle = plt.Circle((xi, yi), radius, color='skyblue', zorder=3)
    ax.add_patch(circle)
    # Label inside the ball
    ax.text(xi, yi, f'$u_{i}$', ha='center', va='center', fontsize=10)

# Draw distance arrows between masses
for i in range(N-1):
    ax.annotate(
        '', 
        xy=(x[i+1]-radius, 0.5), xytext=(x[i]+radius, 0.5),
        arrowprops=dict(arrowstyle='<->', color='red')
    )
    # Label the distance
    ax.text((x[i]+x[i+1])/2, 0.55, 'a', color='red', ha='center', va='bottom', fontsize=10)

# Formatting
ax.set_aspect('equal')
ax.set_xlim(-1, x[-1] + 1)
ax.set_ylim(-1, 1.5)
ax.axis('off')
plt.title("FPU Chain: Masses and Springs with Labels", fontsize=14)
plt.show()
```
